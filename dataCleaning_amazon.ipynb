# Introduction

This project focuses on cleaning and preparing an Amazon dataset downloaded from Kaggle. Real-world datasets often contain inconsistencies such as missing values, duplicates, incorrect data types, and formatting issues that can reduce the quality of analysis. To ensure accurate reporting and reliable insights, a structured ETL (Extract, Transform, Load) process was applied.

* Extract : Raw datasets (customers, orders, deliveries, products, stock) were obtained from Kaggle and loaded into Pandas for processing.

* Transform: Data was cleaned by handling missing values (e.g., discounts, order_status, unit_cost, stock_level), standardizing formats (countries, dates), removing duplicates, and recalculating values using business logic (e.g., margins, delivery times).

* Load: The cleaned datasets are ready to be saved and used for further analysis, visualization, or integration into business intelligence tools.

This project ensures that the Amazon dataset is consistent, accurate, and analysis-ready, forming a solid foundation for exploring customer behavior, sales performance, and operational efficiency.

### Handling Null Values in delivery_date

After inspecting the delivery_date column, I found some missing values. Upon cross-checking with delivery_status, I noticed that some of these orders have already been delivered. Therefore, I will update the missing delivery_date only for orders where delivery_status is “On Time”.

```python
clean_deliveries["delivered"] = np.where(
    ((clean_deliveries["delivery_status"] == "On Time") & clean_deliveries["delivery_date"].isna()) |
    ((clean_deliveries["delivery_status"].isin(["On Time", "Late"])) & clean_deliveries["delivery_date"].notna()),
    1,
    0
)
     
    
clean_deliveries.head(20)
```

```python
clean_deliveries["delivered"].value_counts()
```

#### Findings

After updating the "delivered" column, only 352 rows remain missing.

These missing rows are mostly related to incomplete delivery_date values.

```python
# Convert columns to datetime (if not already)
clean_deliveries['ship_date'] = pd.to_datetime(clean_deliveries['ship_date'], errors='coerce')
clean_deliveries['delivery_date'] = pd.to_datetime(clean_deliveries['delivery_date'], errors='coerce')

# Now calculate delivery time in days
clean_deliveries['delivery_time_days'] = (clean_deliveries['delivery_date'] - clean_deliveries['ship_date']).dt.days

# Check result
clean_deliveries.head(20)
```

```python
## find the average over all time gap 
average_time_all = clean_deliveries["delivery_time_days"].mean()
print("Average delivery day:", average_time_all)
print("-" * 30)


#Average delivery time based on delivery_status
average_time_status = clean_deliveries.groupby("delivery_status")["delivery_time_days"].mean().reset_index()
print("Average delivery time by status:")
print(average_time_status)
print("-" * 30)

average_time_courier = clean_deliveries.groupby("courier")["delivery_time_days"].mean().reset_index()
print("Average delivery time by courier:")
print(average_time_courier)
print("-" * 30)
```

### Overall average delivery time: 3 and half days for all orders.

By delivery status:

Late orders take ~3.50 days on average.

On Time orders take ~3.47 days on average.

By courier:

Amazon Logistics: ~3.52 days

DHL: ~3.51 days

FedEx: ~3.48 days

UPS: ~3.44 days

### Now I will update the delivery_date Nan values  by adding 3 days to handle missing data

```python
clean_deliveries["delivery_date"] = clean_deliveries["delivery_date"].fillna(clean_deliveries["ship_date"] + pd.Timedelta(days=3))
clean_deliveries["delivery_time_days"] = clean_deliveries["delivery_time_days"].fillna(3.0)
clean_deliveries.isna().sum() ## No missing values
```

## Data Cleaning for Order items

```python
order_items.head(10)
```

```python
order_items.count()
```

```python
print("Duplicate order_id :",  order_items["order_id"].duplicated().sum())
```

### Findings

Found missing values in the discount column.

Since discount is directly related to unit_price, quantity, and total_price, I will recalculate the missing values instead of dropping them.
#### Formula to calculate discount :
discount = (unit_price * quantity)- total_price / (unit_price * quantity) 

```python
clean_order_items = order_items.copy()
```

```python
# Create the mask for missing discounts , it will create Bollean Series
mask = clean_order_items["discount"].isnull()
mask.head(10)
```

```python
# Recalculating discount only for missing rows
clean_order_items.loc[mask, "discount"] = (
    (clean_order_items.loc[mask, "unit_price"] * clean_order_items.loc[mask, "quantity"] - clean_order_items.loc[mask,"total_price"])/ (clean_order_items.loc[mask, "unit_price"] * clean_order_items.loc[mask, "quantity"] )).round(2)
```

```python
clean_order_items.isna().sum()
```

### Findings

Found missing values in the order_status column.

Since order_status can be inferred from delivery information, it is possible to recover some of these missing values instead of dropping rows.


```python
clean_orders = orders.copy()
merged_clean_orders = pd.merge(clean_orders, 
                               clean_deliveries[["order_id", "delivered"]], 
                               on="order_id", how="left")

# Fill missing order_starus based on delivered column
mask_orders = merged_clean_orders["order_status"].isnull()

# Fill missing order_status based on delivered column
merged_clean_orders.loc[mask_orders, "order_status"] = merged_clean_orders.loc[mask_orders, "delivered"].map({
    1: "Delivered",
    0: "Shipped"
})
```

```python
#Drop the "delivered" column after filling missing row
merged_clean_orders = merged_clean_orders.drop(columns = ["delivered"])
merged_clean_orders.isna().sum()
```

```python
#Now update missing values with "Unknown"
merged_clean_orders["order_status"] = merged_clean_orders["order_status"].fillna("Unknown")
merged_clean_orders.isna().sum()
```

### Findings

Found missing values in the order_status column.

Instead of dropping rows, these values can be recovered using delivery information (deliveries dataset).

```python
clean_products = products.copy()
```

```python
clean_products['unit_price'] = clean_products['unit_price'].replace('[^\\d.]', '', regex=True)

clean_products['unit_price'] = pd.to_numeric(clean_products['unit_price'], errors='coerce')
```

```python
clean_products['margin_temp'] = (clean_products['unit_price'] - clean_products['unit_cost']) / clean_products['unit_price']
```

```python
# Compute average margin per category (ignoring missing unit_cost)
category_margin = clean_products.loc[clean_products['unit_cost'].notnull()].groupby('category')['margin_temp'].mean()
print("Category average margin:\n", category_margin)
```

### Margin for missing values

Books              0.555274

Electronics        0.482107

Fashion            0.488108

Margin per product (fraction) can be defined as:

margin = (unit_price - unit_cost)/ unit_price


```python
# Fill only the 5 missing unit_cost rows using category margin
mask = clean_products['unit_cost'].isnull() & clean_products['unit_price'].notnull()
clean_products.loc[mask, 'unit_cost'] = clean_products.loc[mask, 'unit_price'] * (1 - clean_products.loc[mask, 'category'].map(category_margin))
```

```python
clean_products['margin'] = (clean_products['unit_price'] - clean_products['unit_cost']) / clean_products['unit_price']
clean_products.drop(columns=['margin_temp'], inplace=True)
```

## Project Summary – Amazon Data Cleaning

In this project, I performed a structured **data cleaning process** across multiple Amazon datasets (`customers`, `orders`, `deliveries`, `products`, `stock`). The goal was to improve **data quality, consistency, and reliability** for further analysis and reporting.



### Key Findings & Actions

1. **Customers Data**

   * Found missing values in `email` and `country`.
   * Standardized country names and filled missing values with `"Unknown"` (for reporting) while keeping critical rows intact.
   * Converted `registration_date` into proper datetime format.
   * Identified and handled duplicates using `customer_id` as the primary key.

2. **Orders Data**

   * Found missing values in `order_status`.
   * Recovered missing statuses by merging with `deliveries` dataset (`delivered` flag).
   * Remaining unknowns were flagged for further investigation instead of being dropped.

3. **Deliveries Data**

   * Missing `delivery_date` values were addressed using the **average delivery time gap** between `ship_date` and `delivery_date`.
   * This ensured realistic delivery timelines without discarding useful rows.

4. **Products Data**

   * Identified missing values in `unit_cost`.
   * Recalculated missing costs using **category-level margins** and cleaned up `unit_price` formatting.
   * Ensured all numeric columns were properly typed (`float` or `int`).

5. **Stock Data**

   * Found missing `stock_level` values.
   * Since averages could not be used, missing values were handled safely (filled with `0` for reporting or left as `NaN` for operational analysis).

6. **Discounts**

   * Missing discounts were recalculated using the formula:

     $$
     \text{discount} = \frac{(\text{unit_price} \times \text{quantity}) - \text{total_price}}{\text{unit_price} \times \text{quantity}}
     $$

   * This ensured consistency between price, quantity, and discount columns.

---

### Outcomes

* All major missing values were either **recovered, filled, or flagged**.
* Duplicate rows were removed where safe.
* Data types were standardized across all datasets.
* Key relationships between datasets (`orders ↔ deliveries`, `products ↔ stock`) were preserved for future analysis.
